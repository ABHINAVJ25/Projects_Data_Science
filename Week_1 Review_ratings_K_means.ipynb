```python
df_n = df.drop(["User"], axis = 1 )
```

```python
df_n.head(5)
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters = 2, n_init="auto")
k_means.fit(df_n)
```

```python
k_means.labels_
```

```python
k_means = KMeans(n_clusters = 2, n_init="auto")
k_means.fit(df_n)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 3, n_init="auto")
k_means.fit(df_n)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 1, n_init="auto")
k_means.fit(df_n)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 4, n_init="auto")
k_means.fit(df_n)
k_means.inertia_
```

```python
wss = []
```

```python
for i in range(1,11):
    KM = KMeans(n_clusters = i, n_init = "auto")
    KM.fit(df_n)
    wss.append(KM.inertia_)
```

```python
wss
```

```python
plt.plot(range(1,11), wss);
```

```python
k_means = KMeans(n_clusters = 2, n_init = "auto")
k_means.fit(df_n)
labels = k_means.labels_
```

```python
df["cluster_kmeans"] = labels
```

```python
df.head(5)
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
silhouette_score(df_n, labels)
```

```python
sil_width = silhouette_samples(df_n, labels)
```

```python
df["sil_width"] = sil_width
```
