## now we are going to use linkage and dendrogram functions included in python's scientific package in order to perform the herirchical clustering 

1. we can use sklearn cluster module which include agglomerative clustering as a function available .
2. by using the scientific pythons package and within that we havea dendrogram and the linkage functions.

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
wardlink = linkage(data, method = "ward")
```

```python
dendrogram(wardlink)
```

```python
dend = dendrogram(wardlink)
```

```python
dend = dendrogram(wardlink,
                 truncate_mode="lastp",
                 p = 10, 
                 )
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
# Method 1 that is what we are talking about
## cluster = fcluster(wardlink, t, criterion = "maxclust" )

cluster = fcluster(wardlink, 3, criterion = "maxclust")  ## passing the variable t = 3
```

```python
cluster
```

```python
# method 2 t = 23 and criterion = "distance"

cluster = fcluster(wardlink, 23, criterion = "distance")
cluster
```

```python
# method 2 just changed the value of the variable t = 22
cluster = fcluster(wardlink, 22, criterion = "distance")
cluster
```

## It happened because three vertical lines are passing by y-axis value 23
    therefore their three clusters present..

```python
df["cluster"] = cluster
```
