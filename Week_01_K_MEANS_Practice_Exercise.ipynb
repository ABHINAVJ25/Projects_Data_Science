## Q4. Standardize the data.
Drop the categorical variable before clustering the data.

```python
data_df = df.drop(["class"], axis = 1)
data_df
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler()
scaled_df = X.fit_transform(data_df)
```

```python
scaled_df
```

```python
from scipy.stats import zscore
scaled_df = data_df.apply(zscore)
scaled_df.head()
```

## Q5. Perform Hierarchical Clustering with the Ward's linkage method and plot the dendrogram.
## Note: Please do go ahead and explore other parameters under the linkage function in the Scientific Python library.

```python
from sklearn.cluster import KMeans
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
wardlink = linkage(scaled_df, method = "ward")
```

```python
dendrogram(wardlink);
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
cluster1 = fcluster(wardlink , 2 , criterion = "maxclust")
```

```python
cluster1
```

```python
cluster2 = fcluster(wardlink, 60, criterion="distance")
```

```python
cluster2
```

```python
np.array_equal(cluster1, cluster2)
```

```python
data_df["cluster"] = cluster1
```

```python
data_df.head()
```

## Q10. Find out the within cluster sum of squares for 2 clusters for the K-Means algorithm.

```python
k_means = KMeans(n_clusters=2, n_init = "auto")
k_means.fit(scaled_df)
k_means.inertia_
```

## Q13. Find the Within Sum of Squares (WSS) for 2 to 15 clusters.

```python
wss = []
```

```python
for i in range(1, 16):
    KM = KMeans(n_clusters = i, n_init = "auto")
    KM = KM.fit(scaled_df)
    wss.append(KM.inertia_)
```

```python
wss
```

## Q14. Plot the Within Sum of Squares (WSS) plot using the values of 'inertia' computed in the last question.

```python
plt.plot(range(1, 16), wss);
```
