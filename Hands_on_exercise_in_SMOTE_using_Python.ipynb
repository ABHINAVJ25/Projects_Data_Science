# Over Sampling using SMOTE

## Import and read data

```python
from numpy import where
import matplotlib.pyplot as plt 

from collections import Counter
from sklearn.datasets import make_classification
from cimport SMOTE
```

```python
# !pip install -U imbalanced-learn
```

## Create an imbalanced data

```python
X, y = make_classification(n_samples = 1000, n_features = 2, n_redundant = 0, n_clusters_per_class = 1, weights = [0.975], flip_y = 0, random_state = 1)
```

```python
counter = Counter(y)
counter
```

```python
for label, _ in counter.items():
    row_ix = where(y == label)[0]
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label = str(label))
plt.legend()
plt.show()
```

```python
## This is im-balanced obviously
```

# OverSampling using SMOTE

```python
oversample = SMOTE()
```

```python
X, y = oversample.fit_resample(X, y)
```

```python
counter = Counter(y)
counter
```

```python
## here we noticed that 
## By resampling the both (X, y) --> and creating more of y = 1, we have created 975 vs 975
```

## And if we visualize this again

```python
for label, _ in counter.items():
    row_ix = where(y == label)[0]
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label = str(label))
plt.legend()
plt.show()
```
